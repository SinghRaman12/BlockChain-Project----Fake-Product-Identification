## PDF && CDF

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
setosa = pd.DataFrame(iris.data[iris.target==0], columns=iris.feature_names)
versicolor = pd.DataFrame(iris.data[iris.target==1], columns=iris.feature_names)
virginica = pd.DataFrame(iris.data[iris.target==2], columns=iris.feature_names)
print(setosa.head())
# getting data of the histogram
count, bins_count = np.histogram(setosa['petal length (cm)'], bins=10)
  
# finding the PDF of the histogram using count values
pdf = count / sum(count)
  
# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)
  
# plotting PDF and CDF
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()

# getting data of the histogram
count, bins_count = np.histogram(versicolor['petal length (cm)'], bins=10)
  
# finding the PDF of the histogram using count values
pdf = count / sum(count)
  
# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)
  
# plotting PDF and CDF
plt.plot(bins_count[1:], pdf, color="blue", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()

# getting data of the histogram
count, bins_count = np.histogram(virginica['petal length (cm)'], bins=10)
  
# finding the PDF of the histogram using count values
pdf = count / sum(count)
  
# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)
  
# plotting PDF and CDF
plt.plot(bins_count[1:], pdf, color="green", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```

# Mean and STD Median

```python
# Compute the mean and standard deviation of petal length for each species
setosa_mean = setosa['petal length (cm)'].mean()
versicolor_mean = versicolor['petal length (cm)'].mean()
virginica_mean = virginica['petal length (cm)'].mean()

setosa_std = setosa['petal length (cm)'].std()
versicolor_std = versicolor['petal length (cm)'].std()
virginica_std = virginica['petal length (cm)'].std()

# Print the results
print("Mean petal length for setosa: {:.2f}".format(setosa_mean))
print("Mean petal length for versicolor: {:.2f}".format(versicolor_mean))
print("Mean petal length for virginica: {:.2f}".format(virginica_mean))

print("Standard deviation of petal length for setosa: {:.2f}".format(setosa_std))
print("Standard deviation of petal length for versicolor: {:.2f}".format(versicolor_std))
print("Standard deviation of petal length for virginica: {:.2f}".format(virginica_std))
```

```python
# Calculate the median of petal length for each species
setosa_median = setosa['petal length (cm)'].median()
versicolor_median = versicolor['petal length (cm)'].median()
virginica_median = virginica['petal length (cm)'].median()

# Print the results
print("Median petal length for setosa: {:.2f}".format(setosa_median))
print("Median petal length for versicolor: {:.2f}".format(versicolor_median))
print("Median petal length for virginica: {:.2f}".format(virginica_median))
```

# Percentile and Quantile

```python
# Calculate the 25th, 50th, and 75th percentiles of petal length for each species
setosa_percentiles = np.percentile(setosa['petal length (cm)'], [25, 50, 75])
versicolor_percentiles = np.percentile(versicolor['petal length (cm)'], [25, 50, 75])
virginica_percentiles = np.percentile(virginica['petal length (cm)'], [25, 50, 75])

# Calculate the first and third quartiles of petal length for each species
setosa_quartiles = np.quantile(setosa['petal length (cm)'], [0.25, 0.75])
versicolor_quartiles = np.quantile(versicolor['petal length (cm)'], [0.25, 0.75])
virginica_quartiles = np.quantile(virginica['petal length (cm)'], [0.25, 0.75])

# Print the results
print("25th, 50th, and 75th percentiles of petal length for setosa: {}".format(setosa_percentiles))
print("25th, 50th, and 75th percentiles of petal length for versicolor: {}".format(versicolor_percentiles))
print("25th, 50th, and 75th percentiles of petal length for virginica: {}".format(virginica_percentiles))

print("First and third quartiles of petal length for setosa: {}".format(setosa_quartiles))
print("First and third quartiles of petal length for versicolor: {}".format(versicolor_quartiles))
print("First and third quartiles of petal length for virginica: {}".format(virginica_quartiles))
```
